```python
import numpy as np
from sklearn.decomposition import PCA

def perform_pca(embeddings_list):
    X = np.array(embeddings_list)

    # Initialize and fit PCA
    pca = PCA()
    reduced_data = pca.fit_transform(X)

    return reduced_data, pca
```

```python
import pandas as pd
import json

def read_json_to_df(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    df = pd.DataFrame(data)
    return df
```

```python
import matplotlib.pyplot as plt

def plot_pca_scatter(reduced_data, df, comp1=0, comp2=1, figsize=(10, 8)):
    """
    Plot PCA scatter plot using specified components

    Parameters:
    reduced_data: The PCA transformed data
    df: Original dataframe with character names/series
    comp1: First component to plot (index)
    comp2: Second component to plot (index)
    figsize: Size of the plot
    """
    # Get coordinates for selected components
    x_coords = reduced_data[:, comp1]
    y_coords = reduced_data[:, comp2]

    # Create plot
    plt.figure(figsize=figsize)

    # Plot points with different colors per series
    for series in df['series'].unique():
        mask = df['series'] == series
        plt.scatter(x_coords[mask], y_coords[mask], label=series, alpha=0.7)

    # Add character names as annotations
    for i, name in enumerate(df.index):
        plt.annotate(name, (x_coords[i], y_coords[i]),
                    xytext=(5, 5), textcoords='offset points',
                    fontsize=8)

    plt.title(f'PCA Components {comp1+1} vs {comp2+1}')
    plt.xlabel(f'Component {comp1+1}')
    plt.ylabel(f'Component {comp2+1}')
    plt.legend()
    plt.grid(True, alpha=0.3)

    return plt.gcf()  # Return the figure object
```

```python
plot_pca_scatter(reduced_data, df, comp1=0, comp2=1)
plt.show()
```

```python
def getVarianceRatio(pca):
    return [f"{x*100:.2f}%" for x in pca.explained_variance_ratio_]

def print_character_coords(reduced_data, pca, df, comp1=0, comp2=1):
   """
   Print character names and their coordinates for two selected PCA components

   Parameters:
   reduced_data: The PCA transformed data
   df: Original dataframe with character names
   comp1, comp2: Component indices to use (0-based)
   """
   x_coords = reduced_data[:, comp1]
   y_coords = reduced_data[:, comp2]
   variance_percentages = getVarianceRatio(pca)

   print(f"Coordinates using Components {comp1+1} and {comp2+1}:")
   print(f"Component {comp1+1} Variance Ratio: {variance_percentages[comp1]}")
   print(f"Component {comp2+1} Variance Ratio: {variance_percentages[comp2]}")
   print(f"Component {comp1+1} is x and {comp2+1} is y")
   print("-" * 50)

   for i, name in enumerate(df.index):
       print(f"{name:20} x: {x_coords[i]:8.3f}, y: {y_coords[i]:8.3f}")
```

```python
import numpy as np

def get_pca_coordinates(data_vectors, pca_component):
    # Returns the coordinate of each vector along the PCA axis
    return np.dot(data_vectors, pca_component) / np.linalg.norm(pca_component)

def analyze_pca_embedding_correlation(pca, original_embeddings):
    """
    Calculate correlation between PCA components and original embedding dimensions.
    Returns the highest correlating embedding dimensions for each PCA component.
    """
    correlations = []
    varianceRatio = getVarianceRatio(pca)

    # For each PCA component
    for i, pca_comp in enumerate(pca.components_):
        # Calculate correlation with each original dimension
        corr_with_dims = []
        coordsAlongComponent = get_pca_coordinates(original_embeddings, pca_comp)
        for j in range(original_embeddings.shape[1]):
            corr = np.corrcoef(coordsAlongComponent, original_embeddings[:, j])[0, 1]
            corr_with_dims.append((j, abs(corr)))  # Store absolute correlation

        # Sort by absolute correlation
        corr_with_dims.sort(key=lambda x: x[1], reverse=True)

        # Get top correlating dimensions
        top_correlations = corr_with_dims[:5]  # Get top 5

        correlations.append({
            'pca_component_index': i,
            'pca_component': pca_comp,
            'variance_by_component': varianceRatio[i],
            'top_correlations': top_correlations,
            'max_correlation': top_correlations[0][1],
            'avg_top5_correlation': np.mean([x[1] for x in top_correlations])
        })

    return correlations

# Example of how to interpret results:
def print_correlation_analysis(correlations):
    for comp in correlations:
        print(f"\nPCA Component {comp['pca_component']}:")
        print(f"Maximum correlation: {comp['max_correlation']:.3f}")
        print(f"Average of top 5 correlations: {comp['avg_top5_correlation']:.3f}")
        print("Top correlating embedding dimensions:")
        print(f"Variance by component {comp['variance_by_component']}")
        for dim, corr in comp['top_correlations']:
            print(f"  Dimension {dim}: {corr:.3f}")
```

```python
def printVarianceByDimension(embeddings):
    varianceByDim = []
    for j in range(embeddings.shape[1]):
        dims = embeddings[:, j]
        variance = np.var(dims)
        varianceByDim.append((j, variance))

    varianceByDim.sort(key=lambda x: x[1], reverse=True)
    totalVariance = sum(map(lambda a: a[1], varianceByDim))
    for i,var in varianceByDim:
        print(f'Dimension {i} variance: {var:.6f}. Percentage of total variance {(var/totalVariance*100):.2f}%')

printVarianceByDimension(embeddingNPArray)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def perform_kmeans(coords, k, random_state=42):
    """
    Returns:
    tuple: (labels, centers, inertia)
        - labels: Cluster assignments for each point
        - centers: Coordinates of cluster centers
        - inertia: Sum of squared distances to nearest cluster center
    """
    kmeans = KMeans(n_clusters=k, random_state=random_state)
    labels = kmeans.fit_predict(coords)
    centers = kmeans.cluster_centers_
    inertia = kmeans.inertia_

    return labels, centers, inertia

def visualize_kmeans_with_labels(coords, labels, centers, point_names, df=None, title="K-means Clustering Results"):
    """
    Visualize k-means clustering results with point annotations

    Parameters:
    coords (array-like): Array of shape (n_samples, 2) containing coordinates
    labels (array): Cluster assignments
    centers (array): Coordinates of cluster centers
    point_names (list): Names to annotate each point
    df (pd.DataFrame, optional): DataFrame containing 'series' column for color coding
    title (str): Plot title
    """
    plt.figure(figsize=(10, 8))

    scatter = plt.scatter(coords[:, 0], coords[:, 1],
                        c=labels, cmap='viridis', alpha=0.7)
    plt.colorbar(scatter, label='Cluster Label')

    # Plot cluster centers
    plt.scatter(centers[:, 0], centers[:, 1],
               c='red', marker='x', s=200, linewidths=3,
               label='Cluster Centers')

    # Add point annotations
    for i, name in enumerate(point_names):
        plt.annotate(name, (coords[i, 0], coords[i, 1]),
                    xytext=(5, 5), textcoords='offset points',
                    fontsize=8)

    plt.title(title)
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
def printPointsAndCluster(df, reduced_data, comp1, comp2, labels, centers, inertia):
    x_coords = reduced_data[:, comp1]
    y_coords = reduced_data[:, comp2]

    for cluster_id in sorted(set(labels)):
        cluster_indices = np.where(labels == cluster_id)[0]

        print(f"\nCluster {cluster_id}:")
        print(f"Center: x={centers[cluster_id][0]:.3f}, y={centers[cluster_id][1]:.3f} | inertia={inertia}")
        print("Points:")
        print("-" * 50)

        for idx in cluster_indices:
            name = df['name'].iloc[idx]
            print(f"{name:20} x: {x_coords[idx]:8.3f}, y: {y_coords[idx]:8.3f}")

        print(f"Total points in cluster: {len(cluster_indices)}")
        print("=" * 50)
```

```python
def plot_elbow(coords, max_k=10):
    inertias = []
    for k in range(1, max_k+1):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(coords)
        inertias.append(kmeans.inertia_)

    plt.plot(range(1, max_k+1), inertias, 'bo-')
    plt.xlabel('k')
    plt.ylabel('Inertia')
    plt.title('Elbow Method')
    plt.show()
```

```python
from sklearn.metrics import silhouette_score

def plot_silhouette(coords, max_k=10):
    scores = []
    for k in range(2, max_k+1):  # Start at 2 since can't compute for k=1
        kmeans = KMeans(n_clusters=k)
        labels = kmeans.fit_predict(coords)
        score = silhouette_score(coords, labels)
        scores.append(score)

    plt.plot(range(2, max_k+1), scores, 'bo-')
    plt.xlabel('k')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Analysis')
    plt.show()

def optimal_k(coords, max_k=10):
    silhouette_scores = []
    for k in range(2, max_k+1):
        kmeans = KMeans(n_clusters=k)
        labels = kmeans.fit_predict(coords)
        score = silhouette_score(coords, labels)
        silhouette_scores.append(score)

    return silhouette_scores.index(max(silhouette_scores)) + 2
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

def plot_dendrogram(X, labels, figsize=(10, 7)):
    """
    Create and plot a dendrogram from the input data.

    Parameters:
    X: array-like of shape (n_samples, n_features)
        The input samples to cluster
    figsize: tuple, default=(10, 7)
        Size of the figure
    """
    # Create linkage matrix using Ward's method
    linkage_matrix = linkage(X, method='ward')

    # Create dendrogram
    plt.figure(figsize=figsize)
    dendrogram(linkage_matrix, labels=labels)
    plt.title('Hierarchical Clustering Dendrogram')
    plt.xlabel('Sample Index')
    plt.ylabel('Distance')
    plt.show()

def perform_hierarchical_clustering(X, n_clusters):
    """
    Perform hierarchical clustering and return labels.

    Parameters:
    X: array-like of shape (n_samples, n_features)
        The input samples to cluster
    n_clusters: int
        The number of clusters to find

    Returns:
    labels: array of shape (n_samples,)
        Cluster labels for each point
    """
    clustering = AgglomerativeClustering(n_clusters=n_clusters)
    labels = clustering.fit_predict(X)
    return labels

def find_optimal_clusters_hierarchical(X, max_clusters=10):
    """
    Find optimal number of clusters using silhouette score.

    Parameters:
    X: array-like of shape (n_samples, n_features)
        The input samples to cluster
    max_clusters: int, default=10
        Maximum number of clusters to test

    Returns:
    dict with optimal number of clusters and scores
    """
    silhouette_scores = []

    # Test different numbers of clusters
    for n_clusters in range(2, max_clusters + 1):
        clustering = AgglomerativeClustering(n_clusters=n_clusters)
        labels = clustering.fit_predict(X)
        score = silhouette_score(X, labels)
        silhouette_scores.append(score)

    # Find optimal number of clusters
    optimal_clusters = silhouette_scores.index(max(silhouette_scores)) + 2

    return {
        'optimal_clusters': optimal_clusters,
        'scores': silhouette_scores,
        'best_score': max(silhouette_scores)
    }

def plot_cluster_comparison(X, max_clusters=6, figsize=(15, 5)):
    """
    Plot data with different numbers of clusters for comparison.

    Parameters:
    X: array-like of shape (n_samples, 2)
        The input samples to cluster (must be 2D for visualization)
    max_clusters: int, default=6
        Maximum number of clusters to show
    figsize: tuple, default=(15, 5)
        Size of the figure
    """
    if X.shape[1] != 2:
        raise ValueError("This visualization only works with 2D data")

    plt.figure(figsize=figsize)

    for i, n_clusters in enumerate(range(2, max_clusters + 1)):
        plt.subplot(1, max_clusters - 1, i + 1)

        # Perform clustering
        clustering = AgglomerativeClustering(n_clusters=n_clusters)
        labels = clustering.fit_predict(X)

        # Plot clusters
        plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
        plt.title(f'{n_clusters} Clusters')

    plt.tight_layout()
    plt.show()
```

```python
# Generate sample data
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=100, n_features=2, centers=4, random_state=42)
# Plot dendrogram
plot_dendrogram(X, [str(i) for i in range(len(X))])

# Find optimal number of clusters
results = find_optimal_clusters_hierarchical(X)
print(f"Optimal number of clusters: {results['optimal_clusters']}")
print(f"Best silhouette score: {results['best_score']:.3f}")

# Visualize different cluster counts
plot_cluster_comparison(X)

# Perform clustering with optimal number of clusters
labels = perform_hierarchical_clustering(X, results['optimal_clusters'])
```
